```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import os
os.chdir('C:/Users/Ahmed Khan/Desktop/Data Sets')
df1 = pd.read_csv("challenger.csv")
df1
```

```python
O_ring_failures = df1['O.ring.failures']
print(O_ring_failures)
```

```python
temp = df1['temperature']
print(temp)
```

```python
plt.plot(temp, O_ring_failures, 'o')
plt.xlabel("Temperature")
plt.ylabel("O_ring_failures")
```

```python
mean_O_ring_failures = O_ring_failures.mean()
print(mean_O_ring_failures)
```

```python
plt.plot(temp, O_ring_failures, 'o')
plt.xlabel("Temperature")
plt.ylabel("O_ring_failures")
plt.axhline(mean_O_ring_failures,color='r',linestyle='-')
```

```python
import statsmodels.api as sm
model = sm.OLS(temp, O_ring_failures).fit()
model.summary()
```

```python
import seaborn as sns
plt.plot(temp, O_ring_failures, 'o')
plt.xlabel("Temperature")
plt.ylabel("O_ring_failures")
plt.axhline(mean_O_ring_failures, color='r', linestyle='-')
sns.regplot(x='temperature', y='O.ring.failures', data=df1)
```

```python
pr = df1['pressure']
plt.plot(pr, O_ring_failures, 'o')
plt.xlabel("Temperature")
plt.ylabel("O_ring_failures")
plt.axhline(mean_O_ring_failures, color='r', linestyle='-')
sns.regplot(x='pressure', y='O.ring.failures', data=df1)
```
